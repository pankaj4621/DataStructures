```python
#check if two vertices have path
class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjacencyMatrix=[[0 for j in range(nVertices)] for i in range(nVertices)]
    
    def addEdge(self,v1,v2):
        self.adjacencyMatrix[v1][v2]=1
        self.adjacencyMatrix[v2][v1]=1
        
    def removeEdge(self,v1,v2):
        if self.isConnected() is False:
            return
        self.adjacencyMatrix[v1][v2]=0
        self.adjacencyMatrix[v2][v1]=0
        
    def isConnected(self,v1,v2):
        if self.adjacencyMatrix[v1][v2]==1:
            return True
        return False
    
    def _hasPathHelper(self,v1,v2,visited):
        if self.adjacencyMatrix[v1][v2]==1:
            return True
        
        visited[v1]=True
        for i in range(self.nVertices):
            if self.adjacencyMatrix[v1][i]==1 and visited[i] is False:
                ans=self._hasPathHelper(i,v2,visited)
                if ans==True:
                    return True
        return False
    
    def hasPath(self,v1,v2):
        visited=[False for i in range(self.nVertices)]
        return self._hasPathHelper(v1,v2,visited)
                
    
    def printGraph(self):
        return self.adjacencyMatrix

    
#main
V,E=input().split()
V=int(V)
E=int(E)
g=Graph(V)
for i in range(E):
    x,y=input().split()
    g.addEdge(int(x),int(y))   
v1,v2=input().split()
result=g.hasPath(int(v1),int(v2))
if result is True:
    print('true')
else:
    print('false')
```

```python
#print all the connected componenets of the graph
from queue import Queue
class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjacencyMatrix=[[0 for j in range(nVertices)] for i in range(nVertices)]
    
    def addEdge(self,v1,v2):
        self.adjacencyMatrix[v1][v2]=1
        self.adjacencyMatrix[v2][v1]=1
        
    def removeEdge(self,v1,v2):
        if self.isConnected() is False:
            return
        self.adjacencyMatrix[v1][v2]=0
        self.adjacencyMatrix[v2][v1]=0
        
    def isConnected(self,v1,v2):
        if self.adjacencyMatrix[v1][v2]==1:
            return True
        return False
    
    def _helper(self,v,visited,l):
        q=Queue()
        q.put(v)
        visited[v]=True
        l.append(v)
        while q.empty() is False:
            k=q.get()
            for j in range(self.nVertices):
                if self.adjacencyMatrix[k][j]==1 and visited[j] is False:
                    q.put(j)
                    visited[j]=True
                    l.append(j)
        return l
    
    def connectedComponent(self):
        ans=[]
        visited=[False for i in range(self.nVertices)]
        for i in range(self.nVertices):
            if visited[i] is False:
                component=self._helper(i,visited,[])
                ans.append(component)
        return ans
    
    def printGraph(self):
        return self.adjacencyMatrix

    
#main
V,E=input().split()
V=int(V)
E=int(E)
g=Graph(V)
for i in range(E):
    x,y=input().split()
    g.addEdge(int(x),int(y))   

ans=g.connectedComponent()
for i in ans:
    i.sort()
    print(*i)
```

```python
#minimum spanning tree (Kruskals Algorithm)
class Edge:
    def __init__(self,src,dest,wt):
        self.src=src
        self.dest=dest
        self.wt=wt
        
def getParent(v,parent):
    if parent[v]==v:
        return v
    return getParent(parent[v],parent)
        
def kruskals(Edges,nVertices):
    parent=[i for i in range(nVertices)]
    Edges=sorted(Edges, key=lambda edge:edge.wt)
    count=0
    output=[]
    i=0
    while count<nVertices-1:
        currEdge=Edges[i]
        srcParent=getParent(currEdge.src,parent)
        destParent=getParent(currEdge.dest,parent)
        if srcParent!=destParent:
            output.append(currEdge)
            parent[srcParent]=destParent
            count+=1
        i+=1
    return output

#main
VandE=[int(i) for i in input().split()]
V=VandE[0]
E=VandE[1]
Edges=[]
for i in range(E):
    currInput=[int(i) for i in input().split()]
    src=currInput[0]
    dest=currInput[1]
    wt=currInput[2]
    edge=Edge(src,dest,wt)
    Edges.append(edge)
    
output=kruskals(Edges,V)
for i in output:
    if i.src<i.dest:
        print(str(i.src)+" "+str(i.dest)+" "+str(i.wt))
    else:
        print(str(i.dest)+" "+str(i.src)+" "+str(i.wt))
    
    
```

```python
#minimum spanning tree (prim's algorithm)

class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjMatrix=[[-1 for j in range(self.nVertices)] for i in range(self.nVertices)]
        
    def addEdge(self,v1,v2,w):
        self.adjMatrix[v1][v2]=w
        self.adjMatrix[v2][v1]=w
        
    def _getMinVertex(self,visited,weight):
        minVertex=-1
        for i in range(self.nVertices):
            if visited[i] is False and (minVertex==-1 or weight[minVertex]>weight[i]):
                minVertex=i
        return minVertex
        
    def prims(self):
        parent=[-1 for i in range(self.nVertices)]
        weight=[float('inf') for i in range(self.nVertices)]
        weight[0]=0
        visited=[False for i in range(self.nVertices)]
        for i in range(self.nVertices-1):
            minVertex=self._getMinVertex(visited,weight)
            visited[minVertex]=True
            for j in range(self.nVertices):
                if self.adjMatrix[minVertex][j]>0 and visited[j] is False:
                    if self.adjMatrix[minVertex][j]<weight[j]:
                        weight[j]=self.adjMatrix[minVertex][j]
                        parent[j]=minVertex
        for i in range(1,self.nVertices):
            if parent[i]>i:
                print(str(i)+" "+str(parent[i])+" "+str(weight[i]))
            else:
                print(str(parent[i])+" "+str(i)+" "+str(weight[i]))
        
        
#main
EandV=[int(i) for i in input().split()]
V=EandV[0]
E=EandV[1]
g=Graph(V)
for i in range(E):
    currInput=[int(i) for i in input().split()]
    g.addEdge(currInput[0],currInput[1],currInput[2])
    
g.prims()
```

```python
#shortest path between source to every other vertex
class Graph:
    def __init__(self,nVertices):
        self.nVertices=nVertices
        self.adjMatrix=[[-1 for j in range(self.nVertices)] for i in range(self.nVertices)]
        
    def addEdge(self,v1,v2,w):
        self.adjMatrix[v1][v2]=w
        self.adjMatrix[v2][v1]=w
        
    def _getMinWeight(self,visited,weight):
        minWeight=-1
        for i in range(self.nVertices):
            if visited[i] is False and (minWeight==-1 or weight[i]<weight[minWeight]):
                minWeight=i
        return minWeight
        
    def dijkstra(self):
        visited=[False for i in range(self.nVertices)]
        weight=[float('inf') for i in range(self.nVertices)]
        weight[0]=0
        
        for i in range(self.nVertices-1):
            minWeight=self._getMinWeight(visited,weight)
            visited[minWeight]=True
            for j in range(self.nVertices):
                if self.adjMatrix[minWeight][j]>0 and visited[j] is False:
                    if weight[j]>weight[minWeight]+self.adjMatrix[minWeight][j]:
                        weight[j]=weight[minWeight]+self.adjMatrix[minWeight][j]
                        
        for i in range(self.nVertices):
            print(str(i)+" "+str(weight[i]))
#main       
EandV=[int(i) for i in input().split()]
V=EandV[0]
E=EandV[1]
g=Graph(V)
for i in range(E):
    currInput=[int(i) for i in input().split()]
    g.addEdge(currInput[0],currInput[1],currInput[2])
g.dijkstra()
```

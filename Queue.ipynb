```python
#defining queue using array
class Queue:
    def __init__(self):
        self.__arr=[]
        self.__count=0
        self.__front=0
        
    def enqueue(self,data):
        self.__arr.append(data)
        self.__count+=1

    def dequeue(self):
        if self.isEmpty():
            return -1
        element=self.__arr[self.__front]
        self.__front+=1
        self.__count-=1
        return element

    def front(self):
        if self.isEmpty():
            return -1
        return self.__arr[self.__front]

    def size(self):
        return self.__count

    def isEmpty(self):
        return self.size()==0 
    
    def lenArray(self):
        return len(self.__arr)

#performing operations on queue
q=Queue()
q.enqueue(10)
q.enqueue(12)
q.enqueue(14)
q.enqueue(16)
while q.isEmpty()==False:
    print(q.front(), "  ""length:",q.lenArray())
    q.dequeue()
print(q.dequeue())
```

```python
#defining queue using linked list
class Node:
    def __init__(self, data):
        self.data=data
        self.next=None
        
class QueueUsingLL:
    def __init__(self):
        self.__head=None
        self.__count=0
    def enqueue(self, data):
        newNode=Node(data)
        if self.__head is None:
            self.__head=newNode
            self.tail=self.__head
            self.__count+=1
            
        else:
            self.tail.next=newNode
            self.tail=self.tail.next
            self.__count+=1
            
    def dequeue(self):
        if self.isEmpty():
            return -1
        self.ele=self.__head.data
        self.__head=self.__head.next
        self.__count-=1
        return self.ele
        
    def front(self):
        if self.isEmpty():
            return -1
        return self.__head.data
    
    def getSize(self):
        return self.__count
    
    def isEmpty(self):
        return self.getSize()==0
    
#performing operations over QueueUsingLL
q = QueueUsingLL()
li = [int(ele) for ele in input().split()]
i=0
while i<len(li):
    choice = li[i]
    if choice == -1:
        break
    elif choice == 1:
        q.enqueue(li[i+1])
        i+=1
    elif choice == 2:
        ans = q.dequeue()
        if ans!=0:
            print(ans)
        else:
            print(-1)
    elif choice == 3:
        ans = q.front()
        if ans!=0:
            print(ans)
        else:
            print(-1)
    elif choice == 4:
        print(q.getSize())
    elif choice == 5:
        if(q.isEmpty()):
            print('true')
        else:
            print('false')
    i+=1
```

```python
#implementation of queue using two stacks Approach 1
#for stacks we are using inbuilt stack i.e list
#enqueue is of O(1)
#dequeue is of O(n)
class Queue2Stack:
    def __init__(self):
        self.__s1=[]
        self.__s2=[]
        self.__count=0
        
    def enqueue(self, data):
        self.__s1.append(data)
        self.__count+=1
        
    def dequeue(self):
        if self.isEmpty():
            return -1
        
        while len(self.__s1)!=1:
            self.__s2.append(self.__s1.pop())
            self.__count-=1
        
        self.ele=self.__s1.pop()
        self.__count-=1
        while self.__s2:
            self.__s1.append(self.__s2.pop())
            self.__count+=1
        return self.ele
    
    def front(self):
        if self.isEmpty():
            return -1
        return self.__s1[0]
        
    def getSize(self):
        return self.__count
    
    def isEmpty(self):
        return self.getSize()==0
    
#performing operations over queue
q=Queue2Stack()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
print("front value: ",q.front())
while not q.isEmpty():
    print(q.dequeue())
print("front value: ",q.front())
```

```python
#implementation of queue using two stacks Approach 2
#for stacks we are using inbuilt stack i.e list
#enqueue is of O(n)
#dequeue is of O(1)
class Queue2Stack:
    def __init__(self):
        self.__s1=[]
        self.__s2=[]
        self.__count=0
        
    def enqueue(self, data):
        while len(self.__s1)!=0:
            self.__s2.append(self.__s1.pop())
            self.__count-=1
        self.__s1.append(data)
        self.__count+=1
        while len(self.__s2)!=0:
            self.__s1.append(self.__s2.pop())
            self.__count+=1
        
    def dequeue(self):
        if self.isEmpty():
            return -1
        self.ele=self.__s1.pop()
        self.__count-=1
        return self.ele
    
    def front(self):
        if self.isEmpty():
            return -1
        return self.__s1[-1]
        
    def getSize(self):
        return self.__count
    
    def isEmpty(self):
        return self.getSize()==0
    
#performing operations over queue
q=Queue2Stack()
q.enqueue(1)
q.enqueue(2)
q.enqueue(3)
q.enqueue(4)
print("front value: ",q.front())
while not q.isEmpty():
    print(q.dequeue())
print("front value: ",q.front())
```

```python
#implementation of stack using two queues
#push O(1)
#pop O(n)
#top o(n)
import queue
class StackUsingQueues:
    
    def __init__(self):
        self.q1 = queue.Queue()
        self.q2 = queue.Queue()
        
        
    def push(self,data):
        self.q1.put(data)
    def pop(self):
        i=1
        n=self.q1.qsize()
        while i<=n-1:
            self.q2.put(self.q1.get())
            i=i+1
        temp=self.q1.get()
        i=1
        while i<=n-1:
            self.q1.put(self.q2.get())
            i=i+1   
        return temp
    def top(self):
        i=1
        n=self.q1.qsize()
        while i<=n-1:
            self.q2.put(self.q1.get())
            i=i+1
        temp=self.q1.get()
        i=1
        while i<=n-1:
            self.q1.put(self.q2.get())
            i=i+1
        self.q1.put(temp)
        return temp
    
    def getSize(self):
        return self.q1.qsize()    
s = StackUsingQueues()
li = [int(ele) for ele in input().split()]
i=0
while i<len(li):
    choice = li[i]
    if choice == -1:
        break
    elif choice == 1:
        s.push(li[i+1])
        i+=1
    elif choice == 2:
        ans = s.pop()
        if ans!=0:
            print(ans)
        else:
            print(-1)
    elif choice == 3:
        ans = s.top()
        if ans!=0:
            print(ans)
        else:
            print(-1)
    elif choice == 4:
        print(s.getSize())
    elif choice == 5:
        while s.q1.qsize() !=0:
            print(s.q1.get(),end=' ')
            
    i+=1
```

```python
#reversing queue without any other queue or stack
import queue
def reverseQueue(q1):
    if q1.qsize()==1:
        return q1
    temp=q1.get()
    reverseQueue(q1)
    q1.put(temp)
    return q1

from sys import setrecursionlimit
setrecursionlimit(11000)
li = [int(ele) for ele in (input().split()[1:])]
q1 = queue.Queue()
for ele in li:
    q1.put(ele)
reverseQueue(q1)
while(q1.empty() is False):
    print(q1.get(),end= ' ')

```

```python
#reversing first k elements of queue in O(n)
import queue
def reverse(q,k):
    if k==0 or q.qsize()==1:
        return 
    temp=q.get()
    reverse(q,k-1)
    q.put(temp)
    return 

def reverseFirstK(q,k):
    reverse(q,k)
    n=q.qsize()
    for i in range(n-k):
        q.put(q.get())
    return
        
from sys import setrecursionlimit
setrecursionlimit(11000)
n = int(input())
li = [int(ele) for ele in input().split()]
q = queue.Queue()
for ele in li:
	q.put(ele)
k = int(input())
reverseFirstK(q,k) 
while(q.qsize()>0):
	print(q.get())
	n-=1
```
